```python
def read_man(file):
    positions = set()
    next(file)
    for line in file:
        line = line.strip().split()
        
        pos = line[1]
        alt = line[4]
        genotype = line[-1]
        
        alt = alt.split(',')
        for gt in genotype.split('|'):
            gt = int(gt)
            if gt != 0:
                positions.add('{}_{}'.format(pos, alt[gt - 1]))
    
    return positions
```

```python
n_people = 30
people: List[Set[int]] = list()
for i in range(n_people):
    people.append(read_man(open(people_dir + f'{i}.vcf')))
```

```python
def find_dist(a, b):
    return -len(a & b)
```

```python
def generate_dists(people):
    n_people = len(people)
    dists = [[0] * n_people for j in range(n_people)]
    for i, man in enumerate(people):
        for j in range(i+1, len(people)):
            dist = find_dist(man, people[j])
            dists[i][j] = dist
            dists[j][i] = dist
    return dists
```

```python
def init(people, n):
    clusters = [list() for _ in range(n)]
    for i in range(len(people)):
        clusters[randrange(n)].append(i)
    return clusters
```

```python
def find_mean(cluster, dists):
    min_man = float('inf'), 0
    for i in cluster:
        dist_all = 0
        for j in cluster:
            dist_all += dists[i][j]
        
        dist_mean = dist_all / len(cluster)
        min_man = min(min_man, (dist_mean, i))
    return min_man[1]
```

```python
def clusterize(people, means, dists):
    n_clusters = len(means)
    n_people = len(people)
    clusters = [list() for _ in range(n_clusters)]
    
    for i in range(n_people):
        min_mean = (float('inf'), 0)
        for m_id, m in enumerate(means):
            dist = (dists[i][m], m_id)
            min_mean = min(min_mean, dist)
        
        min_id = min_mean[1]
        clusters[min_id].append(i)
    
    return clusters
```

```python
def expand_dists(dists, people, man):
    dists.append([0] * (len(dists) + 1))
    print(len(people), len(dists))
    for i in range(len(dists) - 1):
        print(i)
        dist = find_dist(people[i], man)
        dists[i].append(dist)
        dists[-1][i] = dist
```

```python
def knn(man, man_i, clusters, dists, k):
    min_k = [(float('inf'), 0) for _ in range(k)]
    
    for cluster_i, cluster in enumerate(clusters):
        for i in cluster:
            for j, k in enumerate(min_k):
                new_k = (dists[i][-1], cluster_i)
                min_k[j] = min(k, new_k)
    k_counter = [0] * len(clusters)
    for k in min_k:
        k_counter[k[1]] += 1
    relative_cluster = max(enumerate(k_counter), key=lambda x: x[1])[0]
    clusters[relative_cluster].append(man_i)
```
